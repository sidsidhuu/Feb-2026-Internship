1. Unique Words in a Sentence

```python
sentence = "python is easy and python is powerful"
words = sentence.split()
unique_words = set(words)
print("Unique words count:", len(unique_words))
print("Unique words:", unique_words)
```

2. Highest Salary from Employee Data

```python
employees = {
    "Ravi": 75000,
    "Anita": 68000,
    "Kiran": 72000
}

highest_employee = max(employees, key=employees.get)
highest_salary = employees[highest_employee]

print(f"Highest Salary: {highest_employee} - {highest_salary}")
```

3. Find Maximum and Minimum Values

```python
numbers = [45, 22, 89, 10, 66]

maximum = numbers[0]
minimum = numbers[0]

for num in numbers:
    if num > maximum:
        maximum = num
    if num < minimum:
        minimum = num

print(f"List: {numbers}")
print(f"Max: {maximum}")
print(f"Min: {minimum}")
```

4. Count Products Above a Price Threshold

```python
prices = [450, 1200, 899, 1500, 300]

count = 0
for price in prices:
    if price > 1000:
        count += 1

print(f"Products above 1000: {count}")
```

5. Calculate Attendance Percentage

```python
attendance = ["P", "P", "A", "P", "P"]

present_count = attendance.count("P")
total_days = len(attendance)
attendance_percentage = (present_count / total_days) * 100

print(f"Attendance Percentage: {attendance_percentage}")
```

6. Remove Duplicate Phone Numbers

```python
phone_numbers = [9876543210, 9123456789, 9876543210]

unique_phone_numbers = set(phone_numbers)

print(f"Unique phone numbers: {unique_phone_numbers}")
```

7. Count Character Frequency

```python
text = "python"

char_frequency = {}

for char in text:
    if char in char_frequency:
        char_frequency[char] += 1
    else:
        char_frequency[char] = 1

print(char_frequency)
```

8. Convert List to Tuple

```python
my_list = [10, 20, 30]

my_tuple = tuple(my_list)

print(f"Tuple: {my_tuple}")
```

9. Check if a Key Exists in a Dictionary

```python
employees = {"Sita": 50000, "Rahul": 60000}

search_key = "Sita"

if search_key in employees:
    print("Employee exists")
else:
    print("Employee does not exist")
```

10. Calculate Average Marks

```python
marks = [85, 90, 60, 75]

total_marks = sum(marks)
number_of_subjects = len(marks)
average_marks = total_marks / number_of_subjects

print(f"Average Marks: {average_marks}")
```
