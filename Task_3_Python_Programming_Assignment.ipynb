## Program 2: Search Query Keyword Analysis

```python
import stringquery = "Buy mobile phone buy phone online".lower()query = query.translate(str.maketrans('', '', string.punctuation))frequency = {}for word in query.split():    frequency[word] = frequency.get(word, 0) + 1result = {word: count for word, count in frequency.items() if count > 1}print(result)
```
