## Program 1: Employee Performance Bonus Eligibility

```python
employees = {"Ravi": 92, "Anita": 88, "Kiran": 92, "Suresh": 85}
top_score = max(employees.values())
top_performers = [name for name, score in employees.items() if score == top_score]
print(f"Top Performers Eligible for Bonus: {', '.join(top_performers)} (Score: {top_score})")
```

## Program 2: Search Query Keyword Analysis

```python
import string
query = "Buy mobile phone buy phone online".lower()
query = query.translate(str.maketrans('', '', string.punctuation))
frequency = {}
for word in query.split():
    frequency[word] = frequency.get(word, 0) + 1
result = {word: count for word, count in frequency.items() if count > 1}
print(result)
```

## Program 4: Email Domain Usage Analysis

```python
emails = ["ravi@gmail.com", "anita@yahoo.com", "kiran@gmail.com", "suresh@gmail.com", "meena@yahoo.com"]
domain_counts = {}
for email in emails:
    domain = email.split('@')[1]
    domain_counts[domain] = domain_counts.get(domain, 0) + 1
total = len(emails)
for domain, count in domain_counts.items():
    print(f"{domain}: {count / total * 100:.0f}%")
```

## Program 5: Sales Spike Detection

```python
sales = [1200, 1500, 900, 2200, 1400, 3000]
avg_sales = sum(sales) / len(sales)
for day, value in enumerate(sales, start=1):
    if value > avg_sales:
        print(f"Day {day}: {value}")
```

## Program 6: Duplicate User ID Detection

```python
user_ids = ["user1", "user2", "user1", "user3", "user1", "user3"]
counts = {}
for uid in user_ids:
    counts[uid] = counts.get(uid, 0) + 1
for uid, count in counts.items():
    if count > 1:
        print(f"{uid} -> {count} times")
```
