```python
marks = [45, 78, 90, 33, 60]

pass_count = 0
fail_count = 0

for score in marks:
    if score >= 50:
        pass_count += 1
    else:
        fail_count += 1

print(f"Pass: {pass_count}")
print(f"Fail: {fail_count}")
```

```python
names = [" Alice ", "bob", " CHARLIE "]

cleaned_names = [name.strip().lower() for name in names]
print(cleaned_names)
```

```python
messages = ["Hi", "Welcome to the platform", "OK"]

for message in messages:
    length = len(message)
    if length > 10:
        print(f"{message} -> {length} (long)")
    else:
        print(f"{message} -> {length}")
```
